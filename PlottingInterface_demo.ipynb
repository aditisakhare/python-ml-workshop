```python
from bokeh.plotting import figure, output_file, show
import pandas
```

```python
pt = figure(plot_width=650,plot_height=450,title="EarthQuake")
pt.circle([1,2,3,4,5],[4,3,7,9,2],size=[i*5 for i in [4,3,7,9,2]],color="red",alpha=0.5)
```

```python
output_file("scatter_plot.html")
show(pt)
```
